For the left skewness, the mean is less than median; for the right skewness, the mean is greater than median.That's because the right tail for right skewness is bigger, more items accumulate in the bigger values. The other bigger value than the normal distribution makes the mean value become bigger, however, in the same time, median not change. Median is less sensitive than mean.


```python
import plotly.express as px
from scipy import stats
import pandas as pd
import numpy as np

# Generate the right-skewed (positive) sample
sample1 = stats.gamma(a=2, scale=2).rvs(size=1000)
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data", title="Right-Skewed (Positive) Distribution")
fig1.show(renderer="png")  # For GitHub and MarkUs submissions

# Mean and median of sample1 (right-skewed)
mean1 = sample1.mean()
median1 = np.quantile(sample1, 0.5)  # 0.5 quantile is the median

print(f"Sample 1 (Right-Skewed) Mean: {mean1}")
print(f"Sample 1 (Right-Skewed) Median: {median1}")

# Generate the left-skewed (negative) sample by negating the gamma distribution
sample2 = -stats.gamma(a=2, scale=2).rvs(size=1000)
fig2 = px.histogram(pd.DataFrame({'data': sample2}), x="data", title="Left-Skewed (Negative) Distribution")
fig2.show(renderer="png")  # For GitHub and MarkUs submissions

# Mean and median of sample2 (left-skewed)
mean2 = sample2.mean()
median2 = np.quantile(sample2, 0.5)  # 0.5 quantile is the median

print(f"Sample 2 (Left-Skewed) Mean: {mean2}")
print(f"Sample 2 (Left-Skewed) Median: {median2}")
```

from the visualization of the data, we can see that for right skewed distribution(sample 1), the mean is greater than median, and for left skewed distribution, the mean is less than the median.That's there is more values distribute on the left side(left tail) for left skewed distribution than small, makes the mean becomes smaller. But the median is less sensitive, it only depends on the center of the dataset.
